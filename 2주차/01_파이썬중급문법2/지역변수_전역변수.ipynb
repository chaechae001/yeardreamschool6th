# 지역변수 vs 전역변수
- 지역변수(Local) : 함수 안에서 정의 -> 함수 안에서만 존재 (밖에서 접근 불가)
- 전역변수(Global) : 함수 밖에서 정의 -> 내부 프로그램 어디에서나 접근 가능

## 지역 변수
- 함수 안에서만 존재

```python
def greet():
    msg = "함수 안에서 안녕!"  # 지역변수
    print(msg)
greet()
```

```python
print(msg)
```

## 전역변수
- 어디서 접근 가능

```python
msg = "Python is awesome!" # 전역변수

def display():
    print("함수 안:", msg)

display()
```

```python
print("함수 밖:", msg)
```

## Shadowing
- 같은 이름의 지역 변수가 전역변수를 가림

```python
s = "I love yeardreamschool"

def fun():
    s = "Me too."
    print(s)
fun()
```

```python
print(s)
```

# 전역 변수를 수정하고 싶다면?
- global 필요

```python
count = 0
def increase():
    global count # 이 선언이 있으면 전역을 바꿀 수 있음
    count = count + 1
increase()
```

```python
print(count)
```

```python
count = 0
def increase(count=0):
    count = count + 1
increase()
print(count)
```
